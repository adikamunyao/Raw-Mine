### 3.1 Jurisdictional Scope
Encompasses both Spatial and Subject Matter jurisdiction, indicating an IGO’s geographic coverage, thematic focus, influence, and impact.
Jurisdictional scope, covering both spatial and subject matter jurisdiction, is pivotal in global governance. 

`Spatial jurisdiction` delineates geographical areas where an IGO operates, while `subject matter jurisdiction` defines thematic areas of focus. This dual scope clarifies an IGO’s authority and responsibilities, playing a key role in addressing specific environmental and governance issues efficiently. (Berman, 2007; Halliday & Shaffer, 2015; Zürn & Faude, 2013).

#### 3.1.1 Spatial jurisdiction
The spatial jurisdiction of intergovernmental organizations (IGOs) defines the geographic areas over which they exercise authority, particularly within the context of ocean governance. By analyzing these jurisdictions, we aim to assess the overlap and gaps in governance across various maritime zones, such as territorial seas, exclusive economic zones (EEZ), and high seas. This understanding is crucial for identifying opportunities for better coordination and collaboration among organizations involved in ocean governance, ensuring more effective and inclusive management of marine resources.

```python
import altair as alt

# Clean and prepare data for Altair chart
def clean_data(df):
  df['Zones'] = df['Spatial Jurisdiction\n(UNCLOS Zones)'].str.split(', ')
  df_exploded = df.explode('Zones').reset_index(drop=True)
  df_exploded['count'] = 1 
  return df_exploded

df_exploded = clean_data(df.copy())  

# Create zone distribution (specific zones per institution)
zone_distribution = df_exploded.groupby(['Institution'])['Zones'].apply(lambda x: ', '.join(x.unique())).reset_index()

# Count the total jurisdictions and unique zones per institution
institution_stats = df_exploded.groupby('Institution').agg(
    total_jurisdictions=('count', 'sum'),
    unique_zones=('Zones', 'nunique')
).reset_index()

# Merge the statistics with the list of zones
institution_stats = pd.merge(institution_stats, zone_distribution, on='Institution')

# Rename the columns for clarity
institution_stats.columns = ['Institution', 'total_jurisdictions', 'unique_zones', 'specific_zones']

# Create Altair chart
chart = alt.Chart(df_exploded).mark_bar().encode(
    x='Institution:N',
    y=alt.Y('count:Q', title='Number of Jurisdictions'),
    color='Zones:N',
    tooltip=['Institution', 'Zones', 'count']
).properties(
     width=800, height=800,
    title="Spatial Jurisdiction (UNCLOS Zones) by Institution"
).configure_axis(
    labelAngle=-45 
)

# Display chart
chart.show()
print("Organisation Spatial Jurisdiction Statistics")
institution_stats.head()
```

### 3.2 Source of Jurisdiction
The source of jurisdiction for intergovernmental organizations (IGOs) refers to the legal frameworks, treaties, and constitutions that grant these organizations their authority to govern specific areas. For example, the International Oceanographic Commission (IOC) derives its mandate from UNESCO's constitution, while the Food and Agriculture Organization (FAO) is guided by its own constitution and other international agreements like the Convention on Biological Diversity (CBD). Understanding these sources is crucial for determining the legitimacy and scope of each organization’s actions in ocean governance and identifying areas where legal frameworks may need alignment or strengthening.

### Categorization of Jurisdictional Sources in Ocean Governance

#### Introduction
In the realm of ocean governance, various international and regional bodies operate under a diverse set of legal instruments that regulate, manage, and protect ocean resources, maritime boundaries, environmental sustainability, and human rights. These instruments range from binding treaties and agreements to non-legally binding resolutions and declarations. To better understand the complex landscape of ocean governance, this report categorizes the jurisdictional sources based on the **type of legal instrument** they represent, helping facilitate analysis and application across various ocean governance areas.

#### Purpose of Categorization
The objective of this categorization is to simplify the diversity of sources and organize them into broad, meaningful categories based on their legal nature. By focusing on **the type of legal instrument** (e.g., constitutions, statutes, agreements, resolutions), we can better understand how each instrument contributes to the overall governance of oceans, ensuring a more structured and coherent approach to managing ocean-related issues.

#### Methodology
We categorized the sources into the following **types of documents** based on their legal function and role within governance systems. This classification aids in recognizing the diverse instruments at play in ocean governance, from foundational texts to operational guidelines.

- Conventions
- Constitutions
- Statutes
- Treaties
- Charters

These are the primary legal documents that establish the organization and define its scope, powers, and functions. For example, the United Nations is governed by the UN Charter1.

- International Law: This includes customary international law, general principles of law recognized by civilized nations, and judicial decisions. These sources help shape the legal framework within which IGOs operate.
- Agreements
- Protocols

IGOs often enter into various agreements and protocols with member states and other entities to expand their jurisdiction and operational capabilities1.

- Resolutions
- Decisions
The governing bodies of IGOs, such as the UN General Assembly or the Security Council, pass resolutions and make decisions that can have binding effects on member states.

- Administrative Rules and Regulations: These are internal rules and regulations that govern the day-to-day operations of the organization and its interactions with member states and other entities

### Rationale Behind the Categorization
This categorization highlights the diverse legal instruments that play a role in ocean governance. By grouping them based on **document types**—such as statutes, conventions, treaties, resolutions, and others—we can better understand their respective functions and their impact on decision-making, enforcement, and operations in ocean governance.

- **Legal Clarity**: Grouping jurisdictional sources by document type clarifies their enforceability, helping differentiate binding obligations from aspirational or policy-oriented sources.
- **Operational Relevance**: The categories distinguish between high-level governance frameworks (e.g., constitutions and statutes) and the specific rules and regulations that govern activities on the ground (e.g., operational rules, compliance mechanisms).
- **Facilitates Analysis**: This framework simplifies the process of analyzing how various organizations, legal instruments, and agreements interact and complement each other within the broader ocean governance landscape.

By organizing jurisdictional sources in this manner, we provide a clearer structure for understanding the complex system of ocean governance, ultimately supporting more effective analysis and management of ocean-related issues.


```python
import pandas as pd

# Define the categories for documents based on their type
def categorize_document_type(source):
    """
    Function to categorize the document type based on keywords in the source text.
    
    Args:
        source (str): The source text to categorize.
        
    Returns:
        str: The category of the document type (Statute, Constitution, Treaty, Resolution, etc.)
    """
    # Define the keywords associated with each category
    categories = {
        "Statute": ["Statutes", "Charter", "Constitution", "Articles of Agreement", "Statutory"],
        "Treaty": ["Agreement", "Treaty", "Convention", "Protocol", "Part XI", "Memorandum of Understanding"],
        "Resolution": ["Resolution", "General Assembly", "Declaration", "Decision", "Plenary"],
        "Convention": ["Convention", "Treaty", "Agreement", "Protocol"],
        "Regulation": ["Regulation", "Rules", "Administrative Regulations"],
        "Framework": ["Framework", "Programme", "Agenda"],
        "Standard": ["Standard", "Code", "Policy"]
    }
    
    # Initialize a list to store the matched categories
    matched_categories = []
    
    # Loop through each category and check if any keyword is found in the source text
    for category, keywords in categories.items():
        if any(keyword.lower() in source.lower() for keyword in keywords):
            matched_categories.append(category)
    
    # Return the matched categories as a string, or 'Unknown' if no match is found
    return ', '.join(matched_categories) if matched_categories else 'Unknown'


# # Apply the categorize_document_type function to the Primary Source and Other Important Sources
# df['Primary Source Document Type'] = df['Primary Source of Jurisdiction'].apply(categorize_document_type)
# df['Other Sources Document Type'] = df['Other Important Sources'].apply(categorize_document_type)

# # Display the updated DataFrame with document categories
# print(df)

# # Optionally, save to a CSV file for further analysis or reporting
# df.to_csv("jurisdictional_sources_with_categories.csv", index=False)
```

### 3.6 Vertical Coordination
Indicates an IGO’s interactions and collaborations across different levels of governance.

The classification of global institutions in terms of their **vertical integration levels** plays a crucial role in understanding how institutions function within the broader framework of ocean governance. As ocean governance involves a complex web of policy frameworks, agreements, and stakeholders, the need to systematically categorize institutions becomes essential to identify gaps, overlaps, and potential synergies. This report outlines the categorization process, the relevance of vertical integration levels, and how this approach helps in fulfilling the primary objective: assessing and enhancing the effectiveness of global institutions engaged in ocean governance.

#### 2. **Understanding Vertical Integration Levels**

Vertical integration refers to the degree to which an institution or organization is involved in the entire value chain of its operations, from policy formulation to implementation. In the context of ocean governance, vertical integration levels provide insight into how institutions collaborate, implement strategies, and manage resources across different sectors and regions. The classification into different levels of integration helps clarify the roles and influence of institutions at various stages of ocean governance processes.

There are typically three categories for classifying vertical integration:

1. **High Vertical Integration:**
   - Institutions with high vertical integration manage or oversee all aspects of a given domain, including policy formulation, strategic planning, resource management, and implementation.
   - They operate with a broad scope of influence, addressing both global and regional challenges within ocean governance.
   - These institutions are deeply involved in key decision-making processes, providing guidance and technical expertise, while also playing a direct role in enforcement, regulation, and capacity building.

2. **Moderate Vertical Integration:**
   - Institutions with moderate integration focus on specific aspects of ocean governance, such as research, policy advocacy, or specialized technical support.
   - While they may not have direct enforcement or operational authority, they contribute significantly by influencing policy, providing research, and supporting implementation through collaborations and partnerships.
   - They generally play a crucial role in filling gaps or enhancing the capacity of higher-integrated institutions.

3. **Low Vertical Integration:**
   - Institutions in this category typically focus on specific, narrower aspects of ocean governance, with a limited role in policy formulation or operational implementation.
   - Their primary function is to facilitate knowledge-sharing, advocacy, or technical expertise in a focused domain, such as marine biodiversity, pollution control, or fisheries management.
   - They often depend on higher-integrated bodies for policy adoption and implementation.

#### 3. **The Process of Categorization**

The categorization process for vertical integration levels involves several key steps:

1. **Institutional Mapping:**
   - The first step is to identify the core mandate and operational scope of each institution. This involves reviewing their charters, operational frameworks, and strategic priorities.
   - A thorough assessment of each institution’s role in ocean governance, from global policy setting to regional implementation, is conducted.

2. **Assessing Functions and Activities:**
   - Institutions are evaluated based on the following criteria:
     - **Policy Formulation:** Does the institution participate in drafting global treaties or protocols related to ocean governance?
     - **Technical Expertise:** Does it provide scientific, legal, or technical support to governments and other stakeholders?
     - **Implementation:** Is the institution involved in enforcing regulations or directly managing ocean-related activities (e.g., fisheries, pollution)?
     - **Capacity Building:** Does the institution offer training, funding, or other resources to support regional and national efforts?
     - **Coordination:** Does the institution act as a convenor for international dialogues, workshops, or meetings to ensure coordinated actions among stakeholders?

3. **Stakeholder and Partnership Engagement:**
   - Each institution’s degree of collaboration with other stakeholders (governments, NGOs, regional organizations, the private sector) is examined to assess its integration across different levels of governance.
   - Collaboration mechanisms such as joint initiatives, partnerships, and co-authored policy recommendations are reviewed.

4. **Mapping Geographic and Sectoral Reach:**
   - The geographic scope (global or regional) and sectoral focus (e.g., fisheries, biodiversity, pollution control) are considered to understand how institutions operate within the context of ocean governance.
   - Institutions that focus on specific regions or narrower sectors are likely to have a lower level of vertical integration.

5. **Categorization:**
   - Based on these evaluations, each institution is placed into one of the three vertical integration levels: high, moderate, or low. This classification allows us to better understand their respective roles and interconnections in the context of ocean governance.

#### 4. **Relevance to the Objective:**

The categorization of vertical integration levels is directly aligned with the objective of identifying gaps, overlaps, and potential synergies within the global network of institutions involved in ocean governance. By classifying institutions based on their integration levels, we can:

1. **Identify Gaps:**
   - Institutions with low vertical integration might not be effectively addressing certain aspects of ocean governance (e.g., enforcement, capacity building) due to their narrower focus.
   - These gaps may present opportunities for collaboration or the creation of new initiatives that fill the missing links in ocean governance frameworks.

2. **Avoid Overlaps:**
   - High-integrated institutions may have overlapping mandates with other similarly integrated bodies. Understanding these overlaps allows for streamlining processes and reducing inefficiencies by ensuring that responsibilities are clearly defined.
   - This can lead to better resource allocation, improved governance mechanisms, and clearer roles for stakeholders.

3. **Enhance Synergies:**
   - Institutions with moderate integration levels can often bridge the gaps between highly integrated bodies and more specialized, low-integrated institutions.

```python
import pandas as pd
import re

def process_igo_data(raw_text):
    # Define categories for the IGO data
    categories = [
        "Institution Name", 
        "Key Activities/Initiatives", 
        "Collaboration Type", 
        "Collaborating Organizations/Stakeholders", 
        "Key Focus Areas", 
        "Impact/Outcome", 
        "Challenges or Gaps"
    ]
    
    # Define a pattern to capture the relevant sections in the raw text data
    igo_pattern = re.compile(r"(?P<name>[\w\s\-/]+)\s+([\w\s,;.\-:]+)(?=\n|$)")
    
    # Initialize a list to hold processed data
    data = []

    # Loop through each IGO description in the raw text
    for match in igo_pattern.finditer(raw_text):
        igo_name = match.group("name").strip()
        igo_description = match.group(2).strip()

        # Placeholders for each category, can be updated with custom parsing logic
        # Here, we're assuming that the description follows a standard structure
        activities = extract_info(igo_description, "Key Activities/Initiatives")
        collaboration_type = extract_info(igo_description, "Collaboration Type")
        collaborators = extract_info(igo_description, "Collaborating Organizations/Stakeholders")
        focus_areas = extract_info(igo_description, "Key Focus Areas")
        impact_outcome = extract_info(igo_description, "Impact/Outcome")
        challenges = extract_info(igo_description, "Challenges or Gaps")
        
        # Append the processed row data
        data.append([
            igo_name, activities, collaboration_type, collaborators, 
            focus_areas, impact_outcome, challenges
        ])

    # Convert data to a DataFrame for better representation
    df = pd.DataFrame(data, columns=categories)
    return df

def extract_info(text, keyword):
    """ Helper function to extract specific information based on keyword. """
    # A basic logic to find text related to specific keywords, could be improved
    match = re.search(f"{keyword}[:]*([^\n]*)", text)
    return match.group(1).strip() if match else "Not Available"

def main():
    # Sample raw text input (this would be replaced with your full IGO data input)
    raw_text = """
    WTO Joint planning and implementation of programmes and projects with other sectors and stakeholders. 
    Participation in interagency and multisectoral committees, networks, and platforms. 
    Development and dissemination of guidelines, standards, and tools for trade sector engagement with other sectors. 
    Advocacy and communication to raise awareness and mobilize support for trade issues and policies.
    
    UNFPA Coordinate with Member States, UN bodies, and international organizations on sexual and reproductive health, gender equality, and population data.
    Collaborate with governments, civil society, private sector, and academia to implement related policies.
    Support the development and integration of international frameworks and standards.
    Facilitate stakeholder participation in decision-making processes.
    Monitor, assess, and report on progress to inform policies.
    """

    # Process the raw text and generate the table
    igo_df = process_igo_data(raw_text)
    
    # Output the table (e.g., print or save as CSV)
    print(igo_df)

if __name__ == "__main__":
    main()
```
