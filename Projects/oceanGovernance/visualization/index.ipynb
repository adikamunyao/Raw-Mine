```r
# Load necessary libraries
library(ggplot2)

# Prepare the data
data <- data.frame(
  Jurisdiction = c("Global Jurisdiction", 
                   "Member States' Maritime Zones", 
                   "No Specific Authority", 
                   "Varied Jurisdictions"),
  Count = c(37, 9, 1, 1)
)

# Create a bar plot
ggplot(data, aes(x = Jurisdiction, y = Count, fill = Jurisdiction)) +
  geom_bar(stat = "identity", show.legend = FALSE) + # 
  theme_minimal() + # Clean, minimalistic theme
  scale_fill_brewer(palette = "Set2") + 
  labs(
    title = "Distribution of Spatial Jurisdictions",
    x = "Jurisdiction Type",
    y = "Count",
    # caption = "Data Source: Your dataset"
  ) +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1), # Tilt the x-axis labels for readability
    plot.title = element_text(size = 16, face = "bold"), # Title style
    plot.caption = element_text(size = 10, face = "italic") # Caption style
  )
```
